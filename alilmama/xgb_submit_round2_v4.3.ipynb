```python
####################################################### 3.1 线下验证
params={'booster':'gbtree',
    'eta': 0.1,
    'n_estimators':103,
    'max_depth':6, #3 10
    'min_child_weight':4,#1 6
    'gamma':0.3,
    'subsample':0.8,
    'colsample_bytree':0.8,
    'objective': 'binary:logistic',
    'nthread':8,
    'scale_pos_weight':1,
    'eval_metric': 'logloss',
    'lambda':0,
    'seed':0,
    'silent':0}
train_x = filt_feature_data.ix[filt_feature_data["context_timestamp"] < time_stamp_0708, :]
test_x = filt_feature_data.ix[(filt_feature_data["context_timestamp"] >= time_stamp_0708)&
                                                (filt_feature_data["context_timestamp"] < time_stamp_0712), :]
train_x.drop(train_feature_drop, inplace=True, axis=1)
test_x.drop(train_feature_drop, inplace=True, axis=1)
train_y = filt_feature_data.loc[filt_feature_data["context_timestamp"] < time_stamp_0708, ["is_trade"]]
test_y = filt_feature_data.loc[(filt_feature_data["context_timestamp"] >= time_stamp_0708)&
                                                (filt_feature_data["context_timestamp"] < time_stamp_0712), ["is_trade"]]
gc.collect()
dtrain = xgb.DMatrix(data=train_x, label=train_y)
dtest = xgb.DMatrix(data=test_x)

print("xgb start")
del train_x
del train_y
del test_x
gc.collect()

watchlist = [(dtrain,'train')]
bst = xgb.train(params=params, dtrain=dtrain, num_boost_round=200, evals=watchlist, early_stopping_rounds=50)
test_pred = bst.predict(data=dtest)
logloss_prob = metrics.log_loss(y_true=test_y, y_pred=test_pred)
print("logloss_prob:",logloss_prob)
print("xgb build success")
gc.collect()
```
